```python
# Step 3: Load Dataset into Pandas DataFrame
# Assuming you are interested in the demographic data of speakers
df = pd.read_csv('speech_accent_dataset/speakers_all.csv')

# Step 4: Data Cleaning and Preprocessing
#df.dropna(inplace=True)  

# Additional preprocessing steps here as needed...
df['label'], _ = pd.factorize(df['country'])
```

```python
# Define the condition
condition = (df['file_missing?'] == True)
 
# Delete rows based on the condition
df = df[~condition]
```

```python
country_counts = df['country'].value_counts()

# Plotting the bar graph with a larger figure size to accommodate all countries
plt.figure(figsize=(30, 10))  # Adjust the width as needed for 177 countries
country_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Speakers from Each Country')
plt.xlabel('Country')
plt.ylabel('Number of Speakers')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
age_counts = df['age'].value_counts()
plt.figure(figsize=(30, 10))
plt.bar(age_counts.index, age_counts.values)
plt.xlabel('Age')
plt.ylabel('Number of Speakers')
plt.title('Distribution of Ages')
```
